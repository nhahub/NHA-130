## Reading the dataset

```python
csv_path = "../dataset/railway.csv"

dataset = pd.read_csv(csv_path)

dataset
```

## Converting the Dataset into a Dataframe

```python
df = pd.DataFrame(dataset)

df
```

## Handling Missing Values

### Checking for Missing Values

```python
missing_counts = df.isna().sum()
print(missing_counts)
```

### üîç Missing Values Summary

#### Columns with Missing Values:

##### **Railcard**.
- Replace all missing values with `"None"`.

---

##### **Actual Arrival Time**
- Replace all missing values with `"Unknown"`.

---

##### **Reason for Delay**
- If **Journey Status = "Delayed"** and the value is missing ‚Üí set it to `"Unknown"`.
- If the journey is **not delayed**, keep the missing value as is.

---


```python
df['Railcard'] = df['Railcard'].fillna("None")

# Actual Arrival Time
## Replace Missing Values with "Unknown"
df['Actual Arrival Time'] = df['Actual Arrival Time'].fillna("Unknown")

# Reason for Delay
## Replace Missing Values with "Unknown" when 'Journey Status' is "Delayed"
df.loc[
    df[df['Journey Status'] == "Delayed"][df['Reason for Delay'].isna()].index,
    'Reason for Delay'
] = "Unknown"
```

## Data Extraction

### Handling Data Types
Before performing operations on columns, it is important to check and correct their data types.

```python
# Print types of the dataframe
df.dtypes
```

```python
# --- Date columns ---
date_cols = ['Date of Purchase', 'Date of Journey']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# --- Time columns ---
time_cols = ['Time of Purchase', 'Departure Time', 'Arrival Time', 'Actual Arrival Time']
for col in time_cols:
    df[col] = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce').dt.time

# Convert 'Arrival Time' and 'Actual Arrival Time' type to datetime temporarily.
time_cols = ['Arrival Time', 'Actual Arrival Time']
for col in time_cols:
    df[col] = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce')

df
```

### Extracting Day and Month from Date of Journey

```python
df['Day']   = df['Date of Journey'].dt.day
df['Month'] = df['Date of Journey'].dt.month

df
```

### Concatinating Actual Dates with Acutal Times into columns for temporal use

The columns 'Arrival Datetime', 'Actual Arrival Datetime' will be used to generate the delay in minutes by subtracting the actual datetime from the expected one.

```python
df['Arrival Datetime'] = df['Date of Journey'].dt.strftime('%Y-%m-%d') + ' ' + df['Arrival Time'].dt.strftime('%H:%M:%S')
df['Actual Arrival Datetime'] = df['Date of Journey'].dt.strftime('%Y-%m-%d') + ' ' + df['Actual Arrival Time'].dt.strftime('%H:%M:%S')

df['Arrival Datetime'] = pd.to_datetime(df['Arrival Datetime'], errors='coerce')
df['Actual Arrival Datetime'] = pd.to_datetime(df['Actual Arrival Datetime'], errors='coerce')

df
```

### Calculating Delays

```python
mask = df['Actual Arrival Datetime'] < df['Arrival Datetime']
df.loc[mask, 'Actual Arrival Datetime'] += pd.Timedelta(days=1)

df['Delay (min)'] = (df['Actual Arrival Datetime'] - df['Arrival Datetime']).dt.total_seconds() / 60
df.loc[df['Journey Status'] != 'Delayed', 'Delay (min)'] = np.nan

df
```

### Converting 'Arrival Time' and 'Actual Arrival Time' into times

```python
for col in ['Arrival Time', 'Actual Arrival Time']:
    df[col] = pd.to_datetime(df[col], errors='coerce').dt.time

df
```

### Delete the datetimes columns

```python
del df['Arrival Datetime']
del df['Actual Arrival Datetime']

df
```

## Save the cleand data

```python
df.to_csv('../dataset/cleaned_data.csv', index=False)
```
